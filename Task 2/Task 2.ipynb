## Data Cleaning

```python
# Convert price fields (e.g., "$2,100.00") to float.
# price is already in float data type hence there is no need to convert it again.
```

```python
#Parsing dates (e.g., last_review) into datetime objects.
df_listings.info()
```

```python
df_listings['last_review'] = pd.to_datetime(df_listings['last_review'], errors='coerce')
```

```python
df_listings['last_review']
```

```python
# Handling missing values in critical fields like reviews_per_month, host_name, and neighbourhood_group.
df_listings['reviews_per_month'].fillna(0, inplace=True)
df_listings['host_name'].fillna("Unknown", inplace=True)
df_listings['price'].fillna(0, inplace=True)

df_listings.info()
```

```python
# Remove irrelevant rows (e.g., listings with zero availability or price).
df_listings = df_listings[(df_listings['price'] > 0) & (df_listings['availability_365'] > 0 )]
df_listings
```

```python
# Handing neighboorhood_group
df_listings['neighbourhood_group'].fillna("unknown", inplace=True)
df_listings
```

## Data Enrichment

```python
#Create a price_per_booking column using price and minimum_nights.
df_listings['price_per_booking'] = df_listings['price']/df_listings['minimum_nights']
df_listings
```

```python
help
```

```python
# Bucket availability into categories: Full-time (availability > 300) Part-time (100–300) Rare (<100)
df_listings['availability_category'] = pd.cut( 
    df_listings['availability_365'],
    bins=[-1, 100, 300, float('inf')],
    labels=['Rare', 'Part-time', 'Full-Time'],
    include_lowest=True)
df_listings
```

```python
#What are the top 10 most expensive neighborhoods by average price?

##GROUPING THE NEIHBOURHOOD BY AVERAGE PRICE
groupings = df_listings.groupby('neighbourhood')['price'].mean().reset_index().rename(columns={'price': 'av_price'})

##sorting
Top10 = groupings.sort_values(by='av_price', ascending = False).head(10)

print(Top10)
```

```python
# What’s the average availability and price by room type?
av_availability = df_listings.groupby('room_type')[['availability_365', 'price']].mean()
```
